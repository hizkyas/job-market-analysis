```python
import pandas as pd

df = pd.read_csv('../data/jobs_clean.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top 10 locations
top_locations = df['Location'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_locations.values, y=top_locations.index, palette="viridis")
plt.title('Top 10 Job Locations for Data Engineers')
plt.xlabel('Number of Jobs')
plt.ylabel('Location')
plt.show()
```

```python
# Top 10 companies
top_companies = df['Company'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_companies.values, y=top_companies.index, palette="magma")
plt.title('Top 10 Companies Hiring Data Engineers')
plt.xlabel('Number of Jobs')
plt.ylabel('Company')
plt.show()
```

```python
from wordcloud import WordCloud
from collections import Counter

# Skills list counted during cleaning
skills_list = ['Python', 'SQL', 'Spark', 'AWS', 'Docker', 'Kafka', 'Airflow', 'Pandas']
skills_count = Counter()

for title in df['Title']:
    for skill in skills_list:
        if skill.lower() in title.lower():
            skills_count[skill] += 1

# Generate WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(skills_count)

plt.figure(figsize=(15,7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Most In-Demand Skills for Data Engineers")
plt.show()
```

```python
all_words = ' '.join(df['Title']).split()
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(all_words))

plt.figure(figsize=(15,7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("WordCloud of Job Titles")
plt.show()
```
